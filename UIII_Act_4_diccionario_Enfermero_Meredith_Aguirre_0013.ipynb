## Ejemplos de Enfermeria

```python
# 1. Inicialización del diccionario vacío
diccionario_enfermero = {}

# --- FUNCIONES DEL SISTEMA ---

def listar_enfermero():
    print("\n--- LISTA ACTUAL DE ENFERMEROS ---")
    if not diccionario_enfermero:
        print("El diccionario está vacío.")
    else:
        # Formato de tabla manual
        print(f"{'ID':<10} | {'Nombre':<20} | {'Especialidad':<20} | {'Turno':<15}")
        print("-" * 70)
        for clave, datos in diccionario_enfermero.items():
            print(f"{clave:<10} | {datos['Nombre']:<20} | {datos['Especialidad']:<20} | {datos['Turno']:<15}")
    print("-" * 70 + "\n")

def agregar_enfermero():
    id_e = int(input("Ingrese el ID del enfermero: "))
    nombre = input("Ingrese el nombre: ")
    especialidad = input("Ingrese la especialidad: ")
    turno = input("Ingrese el turno: ")

    diccionario_enfermero[id_e] = {
        "Nombre": nombre,
        "Especialidad": especialidad,
        "Turno": turno
    }
    print("\n✅ Enfermero agregado con éxito.")
    listar_enfermero()

def buscar_enfermero():
    id_e = int(input("Ingrese el ID del enfermero a buscar: "))
    if id_e in diccionario_enfermero:
        datos = diccionario_enfermero[id_e]
        print("\n🔍 RESULTADO DE BÚSQUEDA:")
        print(f"{'ID':<10} | {'Nombre':<20} | {'Especialidad':<20} | {'Turno':<15}")
        print("-" * 70)
        print(f"{id_e:<10} | {datos['Nombre']:<20} | {datos['Especialidad']:<20} | {datos['Turno']:<15}")
    else:
        print("\n❌ No se encontró ningún enfermero con ese ID.")
    listar_enfermero()

def borrar_enfermero():
    id_e = int(input("Ingrese el ID del enfermero a eliminar: "))
    if id_e in diccionario_enfermero:
        del diccionario_enfermero[id_e]
        print(f"\n🗑️ Enfermero con ID {id_e} eliminado.")
    else:
        print("\n❌ El ID no existe.")
    listar_enfermero()

def actualizar_enfermero():
    id_e = int(input("Ingrese el ID del enfermero a actualizar: "))
    if id_e in diccionario_enfermero:
        print("Deje en blanco si no desea cambiar el valor.")
        nombre = input(f"Nuevo nombre ({diccionario_enfermero[id_e]['Nombre']}): ")
        especialidad = input(f"Nueva especialidad ({diccionario_enfermero[id_e]['Especialidad']}): ")
        turno = input(f"Nuevo turno ({diccionario_enfermero[id_e]['Turno']}): ")

        if nombre: diccionario_enfermero[id_e]['Nombre'] = nombre
        if especialidad: diccionario_enfermero[id_e]['Especialidad'] = especialidad
        if turno: diccionario_enfermero[id_e]['Turno'] = turno
        print("\n🔄 Datos actualizados.")
    else:
        print("\n❌ ID no encontrado.")
    listar_enfermero()

# --- MENÚ PRINCIPAL ---

while True:
    print("*******************************")
    print("   SISTEMA DE CONTROL MÉDICO   ")
    print("*******************************")
    print("1. Agregar Enfermero")
    print("2. Listar Enfermeros")
    print("3. Buscar Enfermero")
    print("4. Borrar Enfermero")
    print("5. Actualizar Enfermero")
    print("6. Salir")

    opcion = input("\nSeleccione una opción: ")

    if opcion == '1':
        agregar_enfermero()
    elif opcion == '2':
        listar_enfermero()
    elif opcion == '3':
        buscar_enfermero()
    elif opcion == '4':
        borrar_enfermero()
    elif opcion == '5':
        actualizar_enfermero()
    elif opcion == '6':
        print("Saliendo del sistema... ¡Buen turno!")
        break
    else:
        print("Opción no válida, intente de nuevo.")
        print("Programa realizado por Meredith Aguirre NL = 2")
```
